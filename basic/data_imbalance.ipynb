```python
import numpy as np
import matplotlib.pyplot as plt
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
```

```python
X = np.random.randn(80, 2) 
y = [0 for _ in range(80)] # target label = 0
```

```python
X = np.concatenate([X, np.random.randn(20, 2) + np.array([5, 7])], axis=0) # features of label=1
y += [1 for _ in range(20)]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='RdYlGn')
plt.title(f'Original Dataset (class1 ratio={np.mean(y)})')
plt.show()
```

# UnderSampling

```python
rus = RandomUnderSampler()
X_res, y_res = rus.fit_resample(X, y)
```

```python
plt.scatter(X_res[:, 0], X_res[:, 1], c=y_res, cmap='RdYlGn')
plt.title(f'Original Dataset (class1 ratio={np.mean(y_res)})')
plt.show()
```

# OverSampling

```python
ros = RandomOverSampler()
X_res, y_res = ros.fit_resample(X, y)
```

```python
plt.scatter(X_res[:, 0], X_res[:, 1], c=y_res, cmap='RdYlGn')
plt.title(f'Original Dataset (class1 ratio={np.mean(y_res)})')
plt.show()
```

- 초록색에 겹쳐있는 샘플들이 많음

# SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
X = np.random.randn(80, 2)
y = [0 for _ in range(80)]
```

```python
X = np.concatenate([X, np.random.randn(20, 2) + np.array([5, 7])], axis=0) # features of label=1
y += [1 for _ in range(20)]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='RdYlGn')
plt.title(f'Original Dataset (class1 ratio={np.mean(y)})')
plt.show()
```

```python
smote = SMOTE(k_neighbors=7)
X_res, y_res = smote.fit_resample(X, y)
```

```python
plt.scatter(X_res[:, 0], X_res[:, 1], c=y_res, cmap='RdYlGn')
plt.title(f'Original Dataset (class1 ratio={np.mean(y_res)})')
plt.show()
```
